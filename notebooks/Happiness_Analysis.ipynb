```python
# Explore the data
df = pd.read_excel("Happiness.xlsx")
```

```python
# Check for missing values
df.isna().sum()
```

```python
# Sort the dataset by happiness score from highest to lowest
top10 = df.sort_values(by='Score', ascending=False)
```

```python
# Display the first 10 countries
top10.head(10)
```

```python
# Display the country name and happiness score for the top 10 countries
top10[['Country or region', 'Score']].head(10)
```

```python
# Display the 10 least happy countries
bottom10[['Country or region', 'Score']].head(10)
```

```python
# Create a horizontal bar chart for the top 10 happiest countries

plt.figure(figsize=(9,5))

plt.barh(top10['Country or region'].head(10),
         top10['Score'].head(10),
         color='forestgreen',
         edgecolor='black')

plt.title('Top 10 Countries by Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('Happiness Score', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Display the happiest country at the top
plt.gca().invert_yaxis()

# Remove the top and right border
plt.box(False)

plt.show()
```

```python
# Create a horizontal bar chart for the 10 least happy countries

plt.figure(figsize=(9,5))

plt.barh(bottom10['Country or region'].head(10),
         bottom10['Score'].head(10),
         color='firebrick',
         edgecolor='black')

plt.title('Bottom 10 Countries by Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('Happiness Score', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Display the country with the lowest happiness score at the top
plt.gca().invert_yaxis()

# Remove the top and right border
plt.box(False)

plt.show()
```

```python
# Create a scatter plot to compare GDP per capita and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['GDP per capita'],
            df['Score'],
            color='mediumpurple')

plt.title('GDP per Capita vs Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)


plt.show()
```

The scatter plot shows a positive relationship between GDP per capita and happiness score. Countries with a higher GDP per capita generally have higher happiness scores.

```python
# Create a scatter plot to compare social support and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['Social support'],
            df['Score'],
            color='steelblue')

plt.title('Social Support vs Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('Social Support', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)


plt.show()
```

The scatter plot shows a strong positive relationship between social support and happiness score. Countries with higher social support tend to have higher happiness scores.

```python
# Create a scatter plot to compare healthy life expectancy and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['Healthy life expectancy'],
            df['Score'],
            color='mediumseagreen')

plt.title('Healthy Life Expectancy vs Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('Healthy Life Expectancy', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)


plt.show()
```

The scatter plot shows a moderate positive relationship between healthy life expectancy and happiness score. Countries with a longer healthy life expectancy generally have higher happiness scores.

```python
# Create a scatter plot to compare freedom to make life choices and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['Freedom to make life choices'],
            df['Score'],
            color='darkorange')

plt.title('Freedom to make life choices vs Happiness score', fontsize=16, fontweight='bold')
plt.xlabel('Freedom to make life choices', fontsize=12)
plt.ylabel('Happiness score', fontsize=12)


plt.show()
```

The scatter plot shows a moderate positive relationship between freedom to make life choices and happiness score. Countries with more freedom tend to have higher happiness scores.

```python
# Create a scatter plot to compare generosity and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['Generosity'],
            df['Score'],
            color='hotpink')

plt.title('Generosity vs Happiness score', fontsize=16, fontweight='bold')
plt.xlabel('Generosity', fontsize=12)
plt.ylabel('Happiness score', fontsize=12)


plt.show()
```

The scatter plot shows a weak positive relationship between generosity and happiness score. The data suggests that generosity has a smaller effect on happiness.

```python
# Create a scatter plot to compare corruption perception and happiness score

plt.figure(figsize=(8,6))

plt.scatter(df['Perceptions of corruption'],
            df['Score'],
            color='steelblue')

plt.title('Corruption Perception vs Happiness Score', fontsize=16, fontweight='bold')
plt.xlabel('Corruption Perception', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

plt.show()

# The scatter plot shows a moderate positive relationship between corruption perception and happiness score.
# Countries with higher corruption perception scores tend to have higher happiness scores.
```

```python
import seaborn as sns
```

```python
# Calculate the correlation between the numerical columns
correlation = df.corr(numeric_only=True)

# Create a heatmap to show the correlation between variables
plt.figure(figsize=(8,6))

# Display the correlation matrix with correlation values
sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',            # shows correlations with 2 decimal places
            linewidths=0.5)       # adds thin lines between the cells, making the heatmap cleaner and easier to read

# Add a title
plt.title('Correlation Matrix', fontsize=16, fontweight='bold')

# Display the heatmap
plt.show()
```

The heatmap shows the relationship between the numerical variables in the dataset.
It shows that social support and GDP per capita have the strongest positive relationship with happiness score, while generosity has the weakest relationship.
